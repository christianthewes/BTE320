```python
#In class assignment 12
#a
def roi(cost, revenue):
  assert isinstance(cost, (int, float))
  assert isinstance(revenue, (int, float))
  assert cost != 0, "Cost must not be equal to zero"
  return (revenue - cost) / cost

cost = float(input("Enter Cost: "))
revenue = float(input("Enter Revenue: "))
print(roi(cost, revenue))

#b
def loan_payment(principal, annual_rate, months):
  assert (isinstance(principal, (int, float)) and
          isinstance(annual_rate, (int, float)) and
          isinstance(months, (int, float)))
  try:
    monthly_rate = annual_rate / 12
    if monthly_rate == 0:
      monthly_payment = monthly_rate * principal / (1 - (1+monthly_rate) ** (-months))
    else:
      monthly_payment = principal / months
  except:
    raise ValueError("Function inputs must be positive")
```

```python
#In class-assignment 13
def getValues(n):
  assert isinstance(n, int), "Input n must be an integer" # or: assert type(n) == int
  assert n > 0
  weights = []
  for i in range(n):
      value = float(input("Enter a value: "))
      weights.append(value)
  return weights


def calcAverage(alist):
    assert isinstance(alist, list), "Input parameter 'alist' must be a list."
    assert len(alist) > 0
    typical_list = alist[2:]
    return sum(typical_list) / len(typical_list)

    try:
      return sum(typical_list) / len(typical_list)
    except:
      raise ValueError("The length of 'alist' must be greater than 2.")

def main():
  n = int(input("Enter number of items to be added to the list: "))
  if n < 3 and not type(n) == int:
    raise ValueError("The length of 'alist' must be greater than 2.")
  alist = getValues(n)
  average = calcAverage(alist)
  print("Average:", average)

main()
```

```python
#In class Assignment 14
class Greeting:
  def __intin__(self, first, last):
    self.first = first
    self.last = last
  def display(self, formal=False):
    if formal:
      print(f"{self.last}, {self.first}")
    else:
      print(self.first)

  def __str__(self):
    return "Instance of class'Greeting'"

name = Greeting("Jane", "Doe")
print(name)
name.display()
name.display(formal=True)
```

```python
# In-Class Assignment 15

class Student:
  def __init__(self, name, number):
    self.name = name
    self.number = number
    self._courses = []
  def get_name(self):
    return self.name

  def get_courses(self):
    return self._courses

  def set_name(self, newName):
    assert isinstance(newName, str)
    self.name = newName

  def __len__(self):
    return len(self._courses)

  def enroll(self, newCourse):
    if newCourse not in self._courses:
      self._courses.append(newCourse)
    else:
      print(f"{newCourse} is already enrolled in {newCourse}")

s = Student("Jane Doe", 12345)
print(f"Student name: {s.get_name()}")
print(f"Classes already enrolled in: {s.get_courses()}")
s.enroll('BTE 320')
print(f"Classes already enrolled in: {s.get_courses()}")
s.enroll('BTE 320')
print(f"Total number of classes enrolled: {len(s)}")
```

```python
#In class assignment 16
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y
  def distance(self, other):
        dx = (self.x - other.x) ** 2
        dy = (self.y - other.y) ** 2
        return (dx + dy) ** 0.5
class Rectangle:
  def __init__(self):
    self.p1 = Point(0,0)
    self.p2 = Point(5,0)
    self.p3 = Point(0,5)
    self.p4 = Point(5,5)
    self.width = self.p1.distance(self.p2)
    self.height = self.p1.distance(self.p3)
  def area(self):
    return self.width * self.height
```

```python
#In Class Assignment 17
class Employee:
  def __init__(self, name, wage, hours):
    self.name = name
    self.hourly_wage = wage
    self.hours = hours

  def get_wage(self):
    return self.hourly_wage

  def set_wage(self, newWage):
    if newWage <= 0:
      print("Invalid wage value.")
    else:
      self.hourly_wage = new_wage


  def calculate_pay(self):
      return self.hourly_wage * self.hours

  def display_info(self):
    print(f"Employee: {self.name}\nHourly Wage: {self.hourly_wage}")

class Manager(Employee):
  def __init__(self, name, wage, hours, bonus):
    super().__init__(name, wage, hours)
    self.bonus = bonus

  def calculate_pay(self):
    return (1 + self.bonus) * self.get_wage() * self.hours
employee = Employee("Bob", 50, 10)
manager = Manager("Alice", 100, 15, 0.2)

print(employee)
print(manager)
print()
print(f"Employee {employee.name} weekly payment is {employee.calculate_pay()}$")
print(f"Manager {manager.name} weekly payment is {manager.calculate_pay()}$ that includes a {manager.bonus * 100}% bonus")
```
